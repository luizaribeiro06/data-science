```python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 5000

dfCP = pd.DataFrame({
    'ID_Transacao': range(1, n+1),
    'Valor': np.random.uniform(10, 10000, n),
    'Tipo_Transacao': np.random.choice(['Compra', 'Transferencia', 'Pagamento'], n),
    'Localizacao': np.random.choice(['SP', 'RJ', 'MG', 'RS', 'BA', 'PR'], n),
    'Horario': np.random.choice(['Manha', 'Tarde', 'Noite', 'Madrugada'], n),
    'Fraude': np.random.choice([0, 1], n, p=[0.7, 0.3]) #% do 0 de 70% e do 1 30%
})
dfCP.head()
```

```python
#1. Realizar uma amostragem aleatÃ³ria simples com 500 registros.
amostra_simples = dfCP.sample(n=500, random_state=42)
amostra_simples.head()
```

```python
#2. Criar uma amostragem sistemÃ¡tica escolhendo cada 10Âº registro.
amostra_sistematica = dfCP.iloc[::10, :]
amostra_sistematica.head()
```

```python
#3. Dividir a base em estratos por localizaÃ§Ã£o e selecionar amostras proporcionais.
from sklearn.model_selection import train_test_split
amostra_estratificada, _ = train_test_split(dfCP, test_size=0.5, stratify=dfCP["Localizacao"])
amostra_estratificada.head()
```

```python
#4. Selecionar aleatoriamente transaÃ§Ãµes fraudulentas e comparar com transaÃ§Ãµes nÃ£o fraudulentas.
#pegou os que não são fraudulentos
clusters = dfCP.groupby('Fraude')

amostra_nao_fraudulentas = clusters.get_group(0)
amostra_fraudulentas = clusters.get_group(1)

amostra_nao_fraudulentas.head()
amostra_fraudulentas.head()
```

```python
#5. Criar um subconjunto de dados com base em amostragem por julgamento para transaÃ§Ãµes acima de R$5000.
amostra_julgamento = dfCP[(dfCP['Valor'] > 5000)]
amostra_julgamento.head()
```

```python
#6. Aplicar amostragem por conglomerados dividindo os dados por tipo de transaÃ§Ã£o e sorteando um grupo.
import random
clusters = dfCP['Tipo_Transacao'].unique()

sorteio = random.choice(clusters)
grupos = dfCP.groupby('Tipo_Transacao')
amostra_conglomerados = grupos.get_group(sorteio)

amostra_conglomerados.head()
```

```python
#7. Executar uma amostragem por conveniÃªncia pegando os 300 primeiros registros.
amostra_conveniencia = dfCP.head(300)
amostra_conveniencia.head()
```

```python
#8. Criar uma amostragem por cotas considerando o tipo de transaÃ§Ã£o e localizaÃ§Ã£o.
amostra_cotasTransacao = dfCP.groupby('Tipo_Transacao').apply(lambda x: x.sample(n=25)).reset_index(drop=True)
print("Tamanho da amostra por cotas (Tipo de Transação):", len(amostra_cotasTransacao))
amostra_cotasTransacao.head()

amostra_cotasLocalizacao = dfCP.groupby('Localizacao').apply(lambda x: x.sample(n=25)).reset_index(drop=True)
print("\nTamanho da amostra por cotas (Localização):", len(amostra_cotasLocalizacao))
amostra_cotasLocalizacao.head()
```

```python
print("Tamanho da Amostra Aleatória Simples:", len(amostra_simples))
print("Tamanho da Amostra Estratificada:", len(amostra_estratificada))

print("\nConclusão:")
print("Na amostragem simples, os registros são escolhidos de forma totalmente aleatória.")
print("Na estratificada, a escolha é proporcional às localizações (nesse caso), garantindo representatividade.")
```
