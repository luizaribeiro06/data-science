```python
import pandas as pd
import numpy as np

#criando um dataframe fictício
np.random.seed(41)
dados = pd.DataFrame({
    "var_A": np.random.rand(10) * 10, #valores aleatórios entre 0 e 10
    "var_B": np.random.rand(10) * 10, #outra variável aleatória
    "var_C": np.random.rand(10) * 10, #terceira variável aleatória
})

print(dados)

#calculando o coeficiente de correlção de pearson entre A e B
correlacao_AB = dados["var_A"].corr(dados["var_B"])
print(f"Coeficiente de correlação entre A e B: {correlacao_AB:.2f}")
```

```python
#calculando a matriz de correlação para todas as variáveis
matriz_correlacao = dados.corr()
print("\nMatriz de correlação:")
print(matriz_correlacao)
```

```python
#visualizando a matriz de correlação com formatação
import seaborn as sns
import matplotlib.pyplot as plt

#desenha uma figura com esse tamanho
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_correlacao, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5) #escala de 0.5 ponto em 0.5 ponto
plt.title("Matriz de Correlação")
plt.show() #be like print
```
